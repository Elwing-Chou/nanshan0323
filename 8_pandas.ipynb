```python
datas = [
    ["Elwing", 30, "eat"],
    ["Bob", 20, "eat"],
    ["Elwing", 50, "wear"],
    ["Elwing", 100, "eat"]
]
df = pd.DataFrame(datas, columns=["name", "price", "type"])
df.groupby(by="name").sum()
df.groupby(by="name").count()
df.groupby(by=["name", "type"]).sum()
```

```python
datas = [
    ["Elwing", 80],
    ["Bob", 30],
    ["Elwing", 50]
]
df1 = pd.DataFrame(datas, columns=["name", "price"])
datas = [
    ["Elwing", "Taipei", 30],
    ["Amy",  "NY", 20]
]
df2 = pd.DataFrame(datas, columns=["name", "live", "age"])
#  兩邊都有才算
pd.merge(df1, df2, how="inner", on="name")
# 以佐為主
pd.merge(df1, df2, how="left", on="name")
pd.merge(df1, df2, how="right", on="name")
# 合併兩邊的
pd.merge(df1, df2, how="outer", on="name")
```

```python
datas = [
    ["Elwing", 80],
    ["Bob", 30],
    ["Elwing", 50]
]
df1 = pd.DataFrame(datas, columns=["name", "price"])
def change(n):
    return 0.24 * n

df1["price(ja)"] = df1["price"].apply(change)
df1
```

```python
datas = [
    ["Elwing", 80, 90, 79],
    ["Bob", 20, 10, 30]
]
def change(row):
    print(row)
    return row["ch"] * 0.4 + row["math"] * 0.3 + row["en"] * 0.3

df1 = pd.DataFrame(datas, columns=["name", "ch", "math", "en"])
df1.apply(change, axis=0)
```
