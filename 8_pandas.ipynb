```python
data = [
    [1, 2],
    [3, 4]
]
df1 = pd.DataFrame(data, columns=["a", "b"], index=["c", "d"])
df2 = pd.DataFrame(data, columns=["a", "b"], index=["c", "d"])
pd.concat([df1, df2], axis=0)
pd.concat([df1, df2], axis=1)
```

```python
datas = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
df = pd.DataFrame(datas, columns=["a", "b", "c"])
# axis控制直/橫
df.mean()
df.median()
df.var()
df.sum()
```
